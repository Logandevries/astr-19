```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)

random_numbers = np.random.normal(size=1000)
```

```python
plt.hist(random_numbers, bins=100, density=True, alpha=0.6, color='g')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Numbers')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Numbers')

plt.show()
plt.savefig('random_numbers_histogram.pdf')
```
