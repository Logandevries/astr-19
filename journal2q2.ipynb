```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 100)
```

```python
def sin_function(x):
    return np.sin(x)

def cos_function(x):
    return np.cos(x)
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
ax1.plot(x, sin_function(x), color='blue')
ax1.set_title('sin(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')

ax2.plot(x, cos_function(x), color='green')
ax2.set_title('cos(x)')
ax2.set_xlabel('x')
ax2.set_ylabel('cos(x)')

plt.show()
plt.savefig('sine_cosine_plot.pdf')
```
